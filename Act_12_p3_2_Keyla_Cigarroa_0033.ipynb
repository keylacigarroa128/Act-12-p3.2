```python
# Para cuando es Positivo (R)
# Calculando el numero usando if else
# datos de entrada NUM, R
# NÚM >= 0
NUM = float(input("ingrese un numero: "))
if NUM >= 0:
  R = "positivo"
else:
  R = "negativo"
print(f"el numero {NUM} es {R}")
print("Programa realizado por Keyla Paola Cigarroa Palacios NL = 12")
```

```python
# Para cuando es Negativo (R)
# Calculando el numero usando if else
# datos de entrada NUM, R
# NÚM >= 0
NUM = float(input("ingrese un numero: "))
if NUM >= 0:
  R = "positivo"
else:
  R = "negativo"
print(f"el numero {NUM} es {R}")
print("Programa realizado por Keyla Paola Cigarroa Palacios NL = 12")
```
